```python
import matplotlib.pyplot as plt

def read_fid_scores(file_path):
    with open(file_path, 'r') as file:
        return [float(line.strip()) for line in file.readlines()]

acgan_path = 'ACGAN_Output/ACGAN_frechet_distances.txt'
dcgan_path = 'DCGAN_Output_50/DCGAN_frechet_distances.txt'
wgan_path = 'WGAN_Output/WGAN_frechet_distances.txt'

ACGAN_scores = read_fid_scores(acgan_path)
DCGAN_scores = read_fid_scores(dcgan_path)
WGANGP_scores = read_fid_scores(wgan_path)

plt.figure(figsize=(10, 5))
plt.title("FID Scores Comparisons for DCGAN vs WGAN vs ACGAN")
plt.plot(DCGAN_scores, label="DCGAN", color='red')
plt.plot(WGANGP_scores, label="WGAN", color='green')
plt.plot(ACGAN_scores, label="ACGAN", color='blue')
plt.xlabel("Epochs")
plt.ylabel("FID Score")
plt.legend()
plt.savefig('FID_comparison.jpg', format='jpeg', dpi=100, bbox_inches='tight')
plt.show()
```
